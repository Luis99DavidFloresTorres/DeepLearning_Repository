```python
import torch
from torch.utils.data import Dataset, DataLoader

class ChatbotDataset(Dataset):
    def __init__(self, datos, tokenizer, max_length=128):
        self.data = []
        self.tokenizer = tokenizer
        self.max_length = max_length
        for i in datos:
            question, answer = i.split('<SEP>')
            self.data.append((question,answer))
    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        question, answer = self.data[index]
        # Tokenizar y codificar las secuencias
        input_ids = self.tokenizer.encode(question, add_special_tokens=True, max_length=self.max_length, truncation=True,padding='max_length')
        target_ids = self.tokenizer.encode(answer, add_special_tokens=True, max_length=self.max_length, truncation=True,padding='max_length')

        # Convertir a tensores de PyTorch
        input_ids = torch.tensor(input_ids, dtype=torch.long)
        target_ids = torch.tensor(target_ids, dtype=torch.long)

        return {"input_ids": input_ids, "target_ids": target_ids}
```

```python
class ChatbotDataset(Dataset):
    def __init__(self, conversations, tokenizer, max_length=512):
        self.conversations = conversations
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.conversations)

    def __getitem__(self, idx):
        conversation = self.conversations[idx]

        # Tokenizar la conversación
        tokens = self.tokenizer(
            conversation,
            add_special_tokens=True,
            truncation=True,
            padding='max_length',
            max_length=self.max_length,
            return_tensors='pt'
        )

        # Dividir en "inputs" y "targets"
        inputs = tokens['input_ids'][:, :-1]
        targets = tokens['input_ids'][:, 1:]

        return {'input_ids': inputs, 'labels': targets}
```

```python
from transformers import GPT2Tokenizer, GPT2Model
import torch

# Cargar tokenizador y modelo GPT-2
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2Model.from_pretrained("gpt2")

# Texto de entrada
input_text = "Este es un ejemplo para ilustrar el proceso."

# Tokenizar y obtener el last_hidden_state
input_ids = tokenizer.encode(input_text, return_tensors="pt")
outputs = model(input_ids)

# Obtener el last_hidden_state
last_hidden_state = outputs.last_hidden_state

# Supongamos que queremos generar texto utilizando esta representación
# y traducirla a puntos finales. Esto podría implicar el uso de la capa de decodificación
# y generación de texto, pero por simplicidad, aquí simplemente utilizamos el tokenizador inverso.
decoded_text = tokenizer.decode(input_ids[0], skip_special_tokens=True)

# Imprimir resultados
print("Texto de Entrada:", input_text)
print("Last Hidden State:", last_hidden_state)
print("Texto Decodificado:", decoded_text)
```
